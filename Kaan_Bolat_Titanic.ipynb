```python
def değişkenler(variable):
    cat= train_data[variable]
    sayı=cat.value_counts()
    plt.figure(figsize=(9,3))
    plt.bar(sayı.index,sayı)
    plt.xticks(sayı.index,sayı.index.values)
    plt.ylabel("Frekans")
    plt.title(variable)
    plt.show()
    print(variable,sayı)
```

```python
category1 = ["Survived","Sex","Pclass","Embarked","SibSp","Parch"]
```

```python
for i in category1:
    değişkenler(i)
```

```python
category2 = ["Cabin","Name","Ticket"]
for i in category2:
    print(train_data[i].value_counts())
```

```python
def hist(variable):
    plt.figure(figsize=(9,3))
    plt.hist(train_data[variable])
    plt.xlabel(variable)
    plt.ylabel("Frekans")
    plt.title(variable)
    plt.show()
```

```python
sayısal=["Fare","Age","PassengerId"]
for i in sayısal:
    hist(i)
```

```python
# Kabin numarasının ilk harfini çekiyoruz
train_data1["Güverte"] = train_data1["Cabin"].apply(lambda x: x[0] if pd.notna(x) else "U")
```

```python
# Pclass ve Güverte arasındaki ilişkiyi tablo olarak basar
pd.crosstab(train_data1["Güverte"], train_data1["Pclass"])
```

```python
train_data1[["Güverte","Fare"]].groupby(["Güverte"],as_index=False).mean().sort_values(by="Fare",ascending=False)
```

```python
# Güverteye göre gruplayıp, hem Hayatta Kalma oranının hem de Bilet Fiyatının ortalamasını aynı anda alma
train_data1[["Güverte", "Survived", "Fare"]].groupby(["Güverte"], as_index=False).mean().sort_values(by="Fare", ascending=False)
```

```python
# Sadece ilgili 3 sütunu seçip, Güverte ve Survived'a göre gruplama
train_data1[["Güverte", "Survived", "Fare"]].groupby(["Güverte", "Survived"], as_index=False).mean()
```

```python
from IPython.display import display

# Gruplama yapacağımız kategorik değişkenler
kategorik_sutunlar = ["Güverte", "Sex", "Embarked", "SibSp"]

# Ortalamasını alacağımız sayısal değişkenler
sayisal_sutunlar = ["Fare", "Age"]

for sutun in kategorik_sutunlar:
    print(f"\n{'='*50}")
    print(f"{sutun.upper()} ve SURVIVED Kırılımında Ortalamalar")
    print(f"{'='*50}")
    
    # İlgili sütunları seçip, kategori ve Survived'a göre grupluyoruz ve ortalamalarını alıyoruz
    ozet_tablo = train_data1[[sutun, "Survived"] + sayisal_sutunlar].groupby([sutun, "Survived"]).mean()
    
    # Tabloyu ekrana düzgünce basıyoruz
    display(ozet_tablo)
```

```python
# Tüm verinin medyanı ile doldurmak için:
train_data1["Fare"] = train_data1["Fare"].fillna(train_data1["Fare"].median())
```

```python
# Pclass ve Sex'e göre gruplayıp, her grubun kendi yaş medyanını boşluklara atama
train_data1["Age"] = train_data1.groupby(["Pclass", "Sex"])["Age"].transform(lambda x: x.fillna(x.median()))
```
